# Scatter Plot

```python
df1 = df.dropna()
```

```python
df2 = df1.head(50)
```

To create a scatter plot we need not consider all the values since its a very big dataset and the plotting ets messy. So let us consider the first 50 values.

```python
df2
```

Let us consider Age and Post to create a scatter plot. 

```python
df2.plot.scatter(x= 'Age', y ='Post')
```

So, considering the first 50 values to create the scatter plot we can see that most of the arrested people are between the age 20 - 30. Also you can see that the post values are more between 600 - 700 which means Northwest is the place where most people get arrested.  





Now let us consider last 50 values.

```python
df2.plot.scatter(x = 'Age', y = 'Post')
```

When you consider the last 50 values to create the scatter plot we can see that most of the arrested people are between the age 20 - 30. Also you can see that the post values are more between 600 - 800 which means Northwest and western is the place where most people get arrested.

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
x=df2['Age']
y=df2['Post']
z=df2['Arrest']
ax.scatter(x,y,z)
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
plt.show()
```

Here Age is the dependent variable which is in x-axis. Post and Arrest are predictor variable which is in y and z axis respectively. The predictor variable tells us about the average age group of people arrested in their locality and their arrest number is based on that. Thus, the predictor variable conveys the information about the age which is the dependent variable.
